```python
import os
import shutil
import re
from google.colab import drive

print("Montando Google Drive...")
drive.mount('/content/drive')

# --- Configurações ---
# Caminho da pasta no seu Google Drive (AJUSTE SE NECESSÁRIO)
source_dir = '/content/drive/MyDrive/DatasetTCC/Câmera1'

# Caminho base de destino no Colab para o CardboardBoxDataset
# O Dataloader espera que a pasta raiz contenha uma pasta com o nome da classe
dest_base_dir = '/content/UniVAD/data/CardboardBox'
dest_class_dir = os.path.join(dest_base_dir, 'cardboard_box') # Pasta específica da classe

# Intervalo de números de amostra considerados "bons"
good_sample_range = range(14, 25) # range(14, 25) vai de 14 a 24

# --- Escolha das amostras (caixas inteiras) para TREINO ---
# Todas as imagens 'good' dessas amostras irão para 'train/good'
# As imagens 'bad' dessas amostras NÃO serão usadas no treino.
# As imagens das amostras NÃO listadas aqui irão para 'test'.
# Vamos escolher duas amostras boas (ex: 14 e 15) como sugerido.
train_samples = [14, 15]
print(f"Amostras designadas para treino (se 'Good'): {train_samples}")

# Expressão regular para analisar os nomes dos arquivos
# Ajustada para garantir que NÃO HÁ código de localização no final
# Ex: board14.1.jpg ou board2.3.1.jpg, MAS NÃO board24.2.b.jpg
# Grupo 1: Número da amostra (ex: 14)
# Grupo 2: Número do ângulo (ex: 1)
# Grupo 3: Número da foto opcional (ex: .1)
filename_pattern_chao_only = re.compile(r'board(\d+)\.(\d+)(\.\d+)?\.jpg$', re.IGNORECASE)


print(f"Diretório de origem: {source_dir}")
print(f"Diretório base de destino: {dest_base_dir}")

# --- Criação das Pastas de Destino ---
train_good_dir = os.path.join(dest_class_dir, 'train', 'good')
test_good_dir = os.path.join(dest_class_dir, 'test', 'good')
test_bad_dir = os.path.join(dest_class_dir, 'test', 'bad')

# Limpa diretórios antigos se existirem e recria a estrutura
if os.path.exists(dest_class_dir):
    print(f"Limpando diretório de destino antigo: {dest_class_dir}")
    shutil.rmtree(dest_class_dir)

print("Criando estrutura de pastas de destino...")
os.makedirs(train_good_dir, exist_ok=True)
os.makedirs(test_good_dir, exist_ok=True)
os.makedirs(test_bad_dir, exist_ok=True)


# --- Processamento ---
files_copied_train = 0
files_copied_test = 0
files_skipped_location = 0
files_skipped_pattern = 0
files_skipped_other = 0

if not os.path.exists(source_dir):
    print(f"ERRO: Diretório de origem não encontrado: {source_dir}")
    print("Verifique o caminho e tente novamente.")
else:
    print("\nIniciando a organização do dataset para CardboardBoxDataset...")

    for filename in os.listdir(source_dir):
        source_path = os.path.join(source_dir, filename)

        # Processa apenas arquivos .jpg
        if os.path.isfile(source_path) and filename.lower().endswith('.jpg'):
            match = filename_pattern_chao_only.match(filename)

            if match:
                # Extrai o número da amostra
                sample_num = int(match.group(1))

                # Determina a classe (Good/Bad)
                is_good = sample_num in good_sample_range
                status_folder = 'good' if is_good else 'bad'

                # Determina o split (Train/Test)
                # Apenas amostras boas podem ir para 'train/good'
                if sample_num in train_samples and is_good:
                    split_folder = 'train'
                    dest_sub_dir = train_good_dir
                    files_copied_train += 1
                else:
                    split_folder = 'test'
                    dest_sub_dir = test_good_dir if is_good else test_bad_dir
                    files_copied_test += 1

                # Constrói o caminho completo de destino
                dest_path = os.path.join(dest_sub_dir, filename)

                # Copia o arquivo
                try:
                    #print(f"Copiando: {filename} -> {split_folder}/{status_folder}") # Log detalhado
                    shutil.copy2(source_path, dest_path)
                except Exception as e:
                    print(f"ERRO ao copiar {filename}: {e}")
                    # Reverte contadores se a cópia falhar
                    if split_folder == 'train': files_copied_train -= 1
                    else: files_copied_test -= 1
                    files_skipped_other += 1

            else:
                # Verifica se foi pulado por ter código de localização
                # Usando uma regex mais ampla para detectar a localização
                location_check_pattern = re.compile(r'board(\d+)\.(\d+)(\.\d+)?([a-zA-Z]\d*)?\.jpg', re.IGNORECASE)
                location_match = location_check_pattern.match(filename)
                if location_match and location_match.group(4) is not None:
                    files_skipped_location += 1
                    # print(f"INFO: Pulado '{filename}' (não é do chão).") # Log detalhado opcional
                else:
                    print(f"AVISO: Arquivo '{filename}' não corresponde ao padrão 'chão'. Pulando.")
                    files_skipped_pattern += 1
        else:
            # Ignora arquivos não-jpg e diretórios
            if not filename.lower().endswith(('.jpg', '.jpeg')):
                 files_skipped_other += 1

    print("\nOrganização concluída!")
    print(f"Arquivos copiados para 'train/good': {files_copied_train}")
    print(f"Arquivos copiados para 'test' (good+bad): {files_copied_test}")
    print("-" * 20)
    print(f"Arquivos pulados (não eram do 'chão'): {files_skipped_location}")
    print(f"Arquivos pulados (não corresponderam ao padrão 'chão'): {files_skipped_pattern}")
    print(f"Arquivos/Diretórios ignorados (não .jpg, etc.): {files_skipped_other}")
    print(f"Total de arquivos processados no diretório de origem: {files_copied_train + files_copied_test + files_skipped_location + files_skipped_pattern + files_skipped_other}")
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

# --- 1. CONFIGURE AQUI O CAMINHO DA PASTA ---
# Substitua pelo caminho correto onde suas imagens estão salvas.
# Se a pasta `overlaid_results` estiver dentro de uma pasta chamada `results`,
# o caminho provavelmente será algo como: "/content/results/overlaid_results"
folder_path = "results/cardboard_box_sequential/overlaid_results/cardboard_box"
# ---------------------------------------------

# Lista todos os arquivos na pasta que terminam com .png ou .jpg
try:
    image_files = [f for f in sorted(os.listdir(folder_path)) if f.endswith(('.png', '.jpg'))]
except FileNotFoundError:
    print(f"ERRO: A pasta não foi encontrada em '{folder_path}'")
    print("Por favor, verifique se o caminho está correto.")
    image_files = []


# Pega apenas as 10 primeiras imagens
images_to_show = image_files[:10]

if not images_to_show:
    print("Nenhuma imagem encontrada na pasta especificada.")
else:
    print(f"Exibindo as {len(images_to_show)} primeiras imagens de: {folder_path}")

    # Define o tamanho da grade de exibição
    # Para 10 imagens, uma grade de 2 linhas e 5 colunas é ideal
    rows = 2
    cols = 5
    fig, axes = plt.subplots(rows, cols, figsize=(20, 8)) # (largura, altura) em polegadas

    # Aplaina o array de eixos para facilitar a iteração
    axes = axes.flatten()

    # Itera sobre os arquivos de imagem e os exibe
    for i, file_name in enumerate(images_to_show):
        # Constrói o caminho completo para a imagem
        img_path = os.path.join(folder_path, file_name)

        # Abre a imagem
        img = Image.open(img_path)

        # Mostra a imagem no subplot correspondente
        axes[i].imshow(img)
        axes[i].set_title(file_name, fontsize=10) # Adiciona o nome do arquivo como título
        axes[i].axis('off') # Remove os eixos

    # Remove os subplots vazios, caso haja menos de 10 imagens
    for i in range(len(images_to_show), len(axes)):
        axes[i].axis('off')

    # Ajusta o layout para evitar que os títulos se sobreponham
    plt.tight_layout()
    plt.show()
```
