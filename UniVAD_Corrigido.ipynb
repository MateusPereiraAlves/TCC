```python
import os
import shutil
import re
from google.colab import drive

print("Montando Google Drive...")
drive.mount('/content/drive')

# --- Configurações ---
# Caminho da pasta no seu Google Drive (AJUSTE SE NECESSÁRIO)
source_dir = '/content/drive/MyDrive/TCC_Dataset/Câmera1'

# Caminho base de destino no Colab para o CardboardBoxDataset
dest_base_dir = '/content/UniVAD/data/CardboardBox'
dest_class_dir = os.path.join(dest_base_dir, 'cardboard_box') # Pasta específica da classe

# Intervalo de números de amostra considerados "bons"
good_sample_range = range(14, 25) # range(14, 25) vai de 14 a 24

# --- INÍCIO DA MODIFICAÇÃO ---
# Lista EXATA de arquivos para a pasta 'train/good'
# (Os nomes já estão com '_' em vez de '.', como o script fará)
specific_train_good_filenames = [
    'board15_1.jpg',
    'board15_2.jpg',
    'board15_3_2.jpg',
    'board19_1.jpg',
    'board19_2.jpg',
    'board19_3.jpg',
    'board19_4.jpg',
    'board23_3.jpg',
]
print(f"Arquivos específicos designados para 'train/good': {len(specific_train_good_filenames)}")
# --- FIM DA MODIFICAÇÃO ---


# Expressão regular para analisar os nomes dos arquivos originais
# Ajustada para garantir que NÃO HÁ código de localização no final
filename_pattern_chao_only = re.compile(r'board(\d+)\.(\d+)(\.\d+)?\.jpg$', re.IGNORECASE)


print(f"Diretório de origem: {source_dir}")
print(f"Diretório base de destino: {dest_base_dir}")

# --- Criação das Pastas de Destino ---
train_good_dir = os.path.join(dest_class_dir, 'train', 'good')
test_good_dir = os.path.join(dest_class_dir, 'test', 'good')
test_bad_dir = os.path.join(dest_class_dir, 'test', 'bad')

# Limpa diretórios antigos se existirem e recria a estrutura
if os.path.exists(dest_class_dir):
    print(f"Limpando diretório de destino antigo: {dest_class_dir}")
    shutil.rmtree(dest_class_dir)

print("Criando estrutura de pastas de destino...")
os.makedirs(train_good_dir, exist_ok=True)
os.makedirs(test_good_dir, exist_ok=True)
os.makedirs(test_bad_dir, exist_ok=True)


# --- Processamento ---
files_copied_train = 0
files_copied_test = 0
files_skipped_location = 0
files_skipped_pattern = 0
files_skipped_other = 0

if not os.path.exists(source_dir):
    print(f"ERRO: Diretório de origem não encontrado: {source_dir}")
    print("Verifique o caminho e tente novamente.")
else:
    print("\nIniciando a organização do dataset para CardboardBoxDataset...")

    for filename in os.listdir(source_dir):
        source_path = os.path.join(source_dir, filename)

        # Processa apenas arquivos .jpg
        if os.path.isfile(source_path) and filename.lower().endswith('.jpg'):
            match = filename_pattern_chao_only.match(filename)

            if match:
                # Extrai o número da amostra
                sample_num = int(match.group(1))

                # --- Renomeia o arquivo ---
                base_name, extension = os.path.splitext(filename)
                new_base_name = base_name.replace('.', '_')
                new_filename = new_base_name + extension
                # --- Fim da renomeação ---

                # --- Lógica de Destino Modificada ---
                is_good_class = sample_num in good_sample_range

                # 1. Verifica se está na lista específica de TREINO
                if new_filename in specific_train_good_filenames:
                    if is_good_class: # Sanidade: verifica se pertence à classe boa
                        split_folder = 'train'
                        dest_sub_dir = train_good_dir
                        files_copied_train += 1
                    else:
                        print(f"AVISO: Arquivo '{new_filename}' está na lista de treino, mas sample {sample_num} não é 'good'. Pulando.")
                        files_skipped_pattern += 1
                        continue # Pula para o próximo arquivo

                # 2. Se NÃO está na lista de treino, vai para TESTE
                else:
                    split_folder = 'test'
                    if is_good_class:
                        # 'test/good' (Amostras 14-24 que NÃO estão na lista específica)
                        dest_sub_dir = test_good_dir
                        files_copied_test += 1
                    else:
                        # 'test/bad' (Amostras fora do range 14-24)
                        dest_sub_dir = test_bad_dir
                        files_copied_test += 1
                # --- Fim da Lógica de Destino ---


                # Constrói o caminho completo de destino com o NOVO nome
                dest_path = os.path.join(dest_sub_dir, new_filename)

                # Copia o arquivo
                try:
                    #print(f"Copiando: {filename} -> {new_filename} em {split_folder}/{status_folder}") # Log detalhado
                    shutil.copy2(source_path, dest_path)
                except Exception as e:
                    print(f"ERRO ao copiar {filename}: {e}")
                    # Reverte contadores se a cópia falhar
                    if split_folder == 'train': files_copied_train -= 1
                    else: files_copied_test -= 1
                    files_skipped_other += 1

            else:
                # Verifica se foi pulado por ter código de localização
                location_check_pattern = re.compile(r'board(\d+)\.(\d+)(\.\d+)?([a-zA-Z]\d*)?\.jpg', re.IGNORECASE)
                location_match = location_check_pattern.match(filename)
                if location_match and location_match.group(4) is not None:
                    files_skipped_location += 1
                else:
                    print(f"AVISO: Arquivo '{filename}' não corresponde ao padrão 'chão'. Pulando.")
                    files_skipped_pattern += 1
        else:
            # Ignora arquivos não-jpg e diretórios
            if not filename.lower().endswith(('.jpg', '.jpeg')):
                files_skipped_other += 1

    print("\nOrganização concluída!")
    print(f"Arquivos copiados para 'train/good': {files_copied_train} (Arquivos específicos)")
    print(f"Arquivos copiados para 'test' (good+bad): {files_copied_test}")
    print("-" * 20)
    print(f"Arquivos pulados (não eram do 'chão'): {files_skipped_location}")
    print(f"Arquivos pulados (não corresponderam a nenhum critério): {files_skipped_pattern}")
    print(f"Arquivos/Diretórios ignorados (não .jpg, etc.): {files_skipped_other}")
    print(f"Total de arquivos processados no diretório de origem: {files_copied_train + files_copied_test + files_skipped_location + files_skipped_pattern + files_skipped_other}")
```

```python
import os
import shutil

# --- Configurações ---

# 1. Defina o diretório base onde as pastas 'train' e 'test' estão
base_dir = '/content/UniVAD/data/CardboardBox/cardboard_box'

# 2. Defina os prefixos dos arquivos para excluir
# (Baseado no seu script anterior, os arquivos agora são 'board3_1.jpg', etc.)
prefixes_to_delete = [
    'board3_',
    'board9_',
    'board10_',
    'board11_'
]

# --- Fim das Configurações ---

print(f"Iniciando exclusão no diretório: {base_dir}")
print(f"Procurando por arquivos que começam com: {prefixes_to_delete}")

deleted_files_count = 0

# Percorre recursivamente todas as pastas dentro do diretório base (train/good, test/good, test/bad)
for root, dirs, files in os.walk(base_dir):
    for filename in files:
        # Verifica se o nome do arquivo começa com algum dos prefixos
        for prefix in prefixes_to_delete:
            if filename.startswith(prefix):
                file_path = os.path.join(root, filename)
                try:
                    os.remove(file_path)
                    print(f"Excluído: {file_path}")
                    deleted_files_count += 1
                except Exception as e:
                    print(f"ERRO ao excluir {file_path}: {e}")

                # Sai do loop interno assim que encontrar uma correspondência
                break

print("\n--- Concluído ---")
print(f"Total de arquivos excluídos: {deleted_files_count}")
```
